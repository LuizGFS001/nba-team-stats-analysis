1 Liste o nome de todos os times e a média de pontos por jogo da temporada 2023-24, do maior para o menor.


```sql
%sql
SELECT team_name, avg_pts
from nba.silver.team_stats
where season = '2023-24'
ORDER BY avg_pts desc
```

2 Quais times terminaram a temporada 2022-23 com mais de 50 vitórias?

```sql
%sql
SELECT team_name, wins
from nba.silver.team_stats
where wins > 50

```

3 Mostre o nome, temporada e aproveitamento (win_pct) dos times que tiveram aproveitamento acima de 60% em qualquer temporada.

```sql
%sql
select team_name, season, win_pct 
from nba.silver.team_stats
where win_pct > 0.60
ORDER BY win_pct desc 
```

4 Quais foram os 5 times com pior defesa (mais pontos sofridos) na temporada 2022-23?

```sql
%sql
select team_name, avg_pts
from nba.silver.team_stats
ORDER BY avg_pts desc
limit 10
```

5 Liste todos os times que tiveram média de assistências acima de 27 por jogo, em qualquer temporada. Mostre o time, a temporada e o valor.


```sql
%sql
select team_name,
avg_ast
from nba.silver.team_stats
where abs(avg_ast) > 27
```

6 Qual foi a média geral de pontos de todos os times em cada temporada? Ou seja, uma linha para 2022-23 e uma para 2023-24.

```sql
%sql
select avg(avg_pts) as media,
season
from nba.silver.team_stats
GROUP by season
```

7 Quantos times terminaram com saldo positivo (plus_minus maior que zero) em cada temporada?


```sql
%sql
select count(team_name) as teams, season
from nba.silver.team_stats
where plus_minus > 0
GROUP by season
```

8 Qual time teve o maior número total de vitórias considerando as duas temporadas juntas?

```sql
%sql
select team_name, sum(wins)
from nba.silver.team_stats
group by team_name 
order by sum(wins) desc
limit 3
```

9 Liste os times que tiveram média de turnovers acima da média geral da liga na temporada 2023-24.

```sql
%sql
SELECT
team_name,
avg_tov
from nba.silver.team_stats
where season = '2023-24'
and avg_tov > (select avg(avg_tov) from nba.silver.team_stats)
ORDER BY avg_tov desc
```

10 Qual a diferença entre o time que mais pontuou e o que menos pontuou em cada temporada?

```sql
%sql
SELECT
season,
max(avg_pts),
min(avg_pts),
round(max(avg_pts)-min(avg_pts), 1) as diferanca
from nba.silver.team_stats
group by season
```

11 Quais times melhoraram o aproveitamento (win_pct) de 2022-23 para 2023-24? Mostre o nome, os dois aproveitamentos e a diferença.

```sql
%sql
select a.team_name, a.win_pct, b.win_pct,
round(b.win_pct - a.win_pct, 3) as diferenca
from nba.silver.team_stats as a
JOIN nba.silver.team_stats as b
on a.team_name = b.team_name
where a.season = '2022-23' and
b.season = '2023-24'
```

12 Quais times estiveram no top 10 de pontuação nas duas temporadas ao mesmo tempo?

```sql
%sql
select team_name
from nba.silver.team_stats
where season = '2023-24' and ranking <= 10 and team_name in (select team_name
from nba.silver.team_stats
where season = '2022-23' and ranking <= 10)
```

13 Qual time teve a maior queda de pontuação média de uma temporada para a outra?

```sql
%sql
select a.team_name, a.avg_pts, b.avg_pts,
round(b.avg_pts - a.avg_pts, 3) as diferenca
from nba.silver.team_stats as a
join nba.silver.team_stats as b
on  a.team_name = b.team_name
where a.season = '2022-23' and b.season = '2023-24'
and (b.avg_pts - a.avg_pts) < 0
order by diferenca asc
limit 1
```

14 Crie um ranking dos times mais eficientes: some a média de pontos com a média de assistências e subtraia a média de turnovers. Chame essa coluna de "efficiency_score" e ordene do maior para o menor, para a temporada 2023-24.

```sql
%sql
SELECT
  team_name,
  season,
  avg_pts,
  avg_ast,
  avg_tov,
  ROUND(avg_pts + avg_ast - avg_tov, 1) AS efficiency_score
FROM nba.silver.team_stats
WHERE season = '2023-24'
ORDER BY efficiency_score DESC
```

15 Quais times tiveram aproveitamento acima de 50% nas duas temporadas mas nunca apareceram no top 5 de pontuação em nenhuma delas?

```sql
%sql
SELECT team_name
FROM nba.silver.team_stats
WHERE win_pct > 0.5
  AND season = '2022-23'
  AND team_name NOT IN (
    SELECT team_name FROM nba.silver.team_stats
    WHERE ranking <= 5
  )
AND team_name IN (
    SELECT team_name FROM nba.silver.team_stats
    WHERE season = '2023-24' AND win_pct > 0.5
  )
```
